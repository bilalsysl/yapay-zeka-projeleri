```python
df['Title']=df['Name'].str.extract(' ([A-Za-z]+)\.',expand=False)
df['Title'].value_counts()
```

```python
df['Title']=df['Title'].replace(['Ms','Mlle'],'Miss')
df['Title']=df['Title'].replace(['Mme','Countess','Lady','Dona'],'Mrs')
df['Title']=df['Title'].replace(['Dr','Major','Col','Sir','Rev','Jonkheer','Capt','Don'],'Mr')
```

```python
df.groupby('Sex')['Age'].mean()
```

```python
df['Age'].fillna(df.groupby('Title')['Age'].transform('median'),inplace=True)
df.groupby('Title')['Age'].transform('median')
```

```python
del df['Cabin']
```

```python
df['Fare'].fillna(df['Fare'].mean(),inplace=True)
```

```python
df['Family']=df['SibSp']+df['Parch']+1
```

```python
df['Embarked'].fillna('s')
```

```python
df.drop(['PassengerId','Ticket'],axis=1,inplace=True)
```

```python
df_dummies=pd.get_dummies(df,drop_first=True)
```

```python
df_train=df_dummies[:891]
df_test=df_dummies[891:]
```

```python
df_train.shape,df_test.shape
```

```python
del df_train['Survived']
```

```python
y=df1['Survived']
```

```python
x=df_train
```

```python
model=Sequential()
model.add(Dense(12,kernel_initializer='uniform',activation='relu'))
model.add(Dense(8,kernel_initializer='uniform',activation='relu'))
model.add(Dense(1,kernel_initializer='uniform',activation='sigmoid')) # classification iÃ§in son satÄ±r bÃ¶yle olmalÄ±
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy']) # 2 cevapli classification iÃ§in binary_crossentropy
# 3 cevaplÄ± classifaction iÃ§in category_crossentropy
```

```python
model.fit(x,y, epochs=150,batch_size=10,verbose=1)
```

```python
model.summary()
```

```python
scores=model.evaluate(x,y)
```
