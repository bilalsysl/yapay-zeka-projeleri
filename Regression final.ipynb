```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
source1=requests.get('https://www.hepsiemlak.com/istanbul-kiralik').text
source2=requests.get('https://www.hepsiemlak.com/istanbul-kiralik?page=2').text
source3=requests.get('https://www.hepsiemlak.com/istanbul-kiralik?page=3').text
source4=requests.get('https://www.hepsiemlak.com/istanbul-kiralik?page=4').text
source5=requests.get('https://www.hepsiemlak.com/istanbul-kiralik?page=5').text
source6=requests.get('https://www.hepsiemlak.com/istanbul-kiralik?page=6').text
source7=requests.get('https://www.hepsiemlak.com/istanbul-kiralik?page=7').text
source8=requests.get('https://www.hepsiemlak.com/istanbul-kiralik?page=8').text
source9=requests.get('https://www.hepsiemlak.com/istanbul-kiralik?page=9').text
source10=requests.get('https://www.hepsiemlak.com/istanbul-kiralik?page=10').text
source=source1+source2+source3+source4+source5+source6+source7+source8+source9+source10
```

```python
soup=BeautifulSoup(source,'html.parser')
```

```python
soup.beautify

  
```

```python
container=soup.find_all("ul",{"class":"list-items-container"})
```

```python
fiyat=soup.findAll("span",{"class":"list-view-price"})
yas=soup.findAll("span",{"class":"celly buildingAge"})
kat=soup.findAll("span",{"class":"celly floortype"})
boyut=soup.findAll("span",{"class":"celly houseRoomCount"})
alan=soup.findAll("span",{"class":"celly squareMeter list-view-size"})
```

```python
f=[]
y=[]
k=[]
b=[]
a=[]
```

```python
for span in fiyat:
    print(span.text)
    f.append(span.text)
```

```python
for span in yas:
    print(span.text)
    y.append(span.text)
```

```python
for span in kat:
    print(span.text)
    k.append(span.text)
```

```python
for span in boyut:
    print(span.text)
    b.append(span.text)
```

```python
for span in alan:
    print(span.text)
    a.append(span.text)
```

```python
d={'fiyat':f,'yas':y,'kat':k,'boyut':b,'alan':a}
df=pd.DataFrame(d)
```

```python
df
```

```python
df.kat.value_counts()
```

```python
df['fiyat']=df['fiyat'].str.replace('\n','')
df['kat']=df['kat'].str.replace('\n','')
df['alan']=df['alan'].str.replace('\n','')
df['fiyat']=df['fiyat'].str.replace('TL','')
df['yas']=df['yas'].str.replace('Yaşında','')
df['yas']=df['yas'].str.replace('Sıfır Bina','0')


df['alan']=df['alan'].str.replace('m2','')
df['fiyat']=df['fiyat'].str.replace('.','')
df['alan']=df['alan'].str.replace('.','')




df['fiyat']=df['fiyat'].astype(int)
df['yas']=df['yas'].astype(int)

df['alan']=df['alan'].astype(int)
```

```python
df.to_csv('istanbul.csv')
```

```python
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2,random_state=13)
```

```python
gbr=GradientBoostingRegressor()
```

```python
öğren=gbr.fit(x_train,y_train)
```

```python
tahmin=öğren.predict(x_test)
```

```python
(mean_squared_error(y_test,tahmin))**0.5
```

```python
(r2_score(y_test,tahmin))
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
regmodel=lr.fit(x_train,y_train) 
```

```python
y_pretest=regmodel.predict(x_test) 
```

```python
(mean_squared_error(y_test,y_pretest))**0.5
```

```python
r2_score(y_test,y_pretest)
```

```python
from sklearn.linear_model import Ridge,Lasso
```

```python
R=Ridge()
R.fit(x_train,y_train)
y_pred_ridge=R.predict(x_test)
```

```python
(mean_squared_error(y_test,y_pred_ridge))**0.5
```

```python
r2_score(y_test,y_pred_ridge) ## en başarılı
```

```python
L=Lasso()
L.fit(x_train,y_train)
y_pred_lasso=L.predict(x_test)
```

```python
(mean_squared_error(y_test,y_pred_lasso))**0.5
```

```python
r2_score(y_test,y_pred_lasso)  
```
