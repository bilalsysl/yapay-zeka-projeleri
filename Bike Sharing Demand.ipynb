```python
df1['datetime']=pd.to_datetime(df1['datetime'])
df1['YÄ±l']=df1['datetime'].dt.year
df1['Ay']=df1['datetime'].dt.month
df1['Gun']=df1['datetime'].dt.dayofweek
df1['Saat']=df1['datetime'].dt.hour
df1['Dakika']=df1['datetime'].dt.minute
```

```python
df1.drop(['datetime','casual','registered'],axis=1,inplace=True)
```

```python
df2['datetime']=pd.to_datetime(df2['datetime'])
df2['YÄ±l']=df2['datetime'].dt.year
df2['Ay']=df2['datetime'].dt.month
df2['Gun']=df2['datetime'].dt.dayofweek
df2['Saat']=df2['datetime'].dt.hour
df2['Dakika']=df2['datetime'].dt.minute
```

```python
h=df2['datetime']
```

```python
df2.drop('datetime',axis=1,inplace=True)
```

```python
x=df1.drop('count',axis=1)
y=df1['count']
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
x_train,x_test,y_train,y_test=tts(x,y,test_size=0.25,random_state=42)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
gbr=GradientBoostingRegressor()
gbr.fit(x_train,y_train)
y_pred=gbr.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
r2_score(y_test,y_pred)
```

```python
(mean_squared_error(y_test,y_pred))**0.5
```
