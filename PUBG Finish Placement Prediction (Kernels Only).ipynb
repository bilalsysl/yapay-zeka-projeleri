```python
df1['winPlacePerc'].fillna(df1['winPlacePerc'].mean(),inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.ensemble import GradientBoostingRegressor
```

```python
df1.drop(['Id','groupId','matchId'],axis=1,inplace=True)
df1=pd.get_dummies(df1)
```

```python
x=df1.drop(['winPlacePerc'],axis=1)
y=df1['winPlacePerc']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2,random_state=42)
```

```python
gbr=GradientBoostingRegressor()
```

```python
öğren=gbr.fit(x_train,y_train)
```

```python
tahmin=öğren.predict(x_test)
```

```python
(mean_squared_error(y_test,tahmin))**0.5
```

```python
(r2_score(y_test,tahmin))
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
regmodel=lr.fit(x_train,y_train) 
```

```python
y_pretest=regmodel.predict(x_test) 
```

```python
(mean_squared_error(y_test,y_pretest))**0.5
```

```python
r2_score(y_test,y_pretest)
```
