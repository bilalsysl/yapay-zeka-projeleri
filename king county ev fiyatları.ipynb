```python
import pandas as pd
```

```python
df=pd.read_csv('kc_house_data.csv')
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df['zipcode']=df['zipcode'].astype('category')
```

```python
df.tail()
```

```python
df.describe()
```

```python
df=df[df['bathrooms']<4]
```

```python
df=df[df['bedrooms']<10]
```

```python
df=df[df['yr_renovated']<1000]
```

```python
df=df[df['yr_built']>1950]
```

```python
df['bedrooms']=df['bedrooms']**2
df['bathrooms']=df['bathrooms']**2
```

```python
df['age']=2021-df['yr_built']
```

```python
df.info()
```

```python
df.shape
```

```python
df.corr()
```

```python
abs(df.corr()['price']).sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
ax=plt.subplots(figsize=(14,12))
sns.heatmap(df.corr(),annot=True)
```

```python
x=df[['sqft_living','grade','sqft_above','sqft_living15','bathrooms','view','sqft_basement','bedrooms','lat','waterfront','floors']]
```

```python
x.head()
```

```python
y=df[['price']]
```

```python
y.shape
```

```python
y.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
lr=LinearRegression()
regmodel=lr.fit(x_train,y_train)       #x train verisine bakıp y train verisini öğrenior
```

```python
y_pretest=regmodel.predict(x_test)      # öğrendiği model ile test için ayırdığımız verinin fiyatlarını tahmin edior
y_pretest
```

```python
from sklearn.metrics import mean_squared_error, r2_score          #r2 %80 üzeri ise başarılı
```

```python
(mean_squared_error(y_test,y_pretest))**0.5
```

```python
r2_score(y_test,y_pretest)
```

```python
x,y=df[['sqft_living','grade','sqft_above','sqft_living15','bathrooms','view','sqft_basement','bedrooms','lat','waterfront','floors','yr_renovated','sqft_lot','zipcode','age']],df[['price']]
```

```python
x=pd.get_dummies(x,drop_first=True)
```

```python
x.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.5, random_state=1)
```

```python
lr=LinearRegression()
regmodel=lr.fit(x_train,y_train)
y_pretest=regmodel.predict(x_test)
y_pretest
```

```python
(mean_squared_error(y_test,y_pretest))**0.5
```

```python
r2_score(y_test,y_pretest)
```

```python
from sklearn.linear_model import Ridge,Lasso
```

```python
R=Ridge()
R.fit(x_train,y_train)
y_pred_ridge=R.predict(x_test)
```

```python
y_pred_ridge
```

```python
(mean_squared_error(y_test,y_pred_ridge))**0.5
```

```python
r2_score(y_test,y_pred_ridge)
```

```python
L=Lasso()
L.fit(x_train,y_train)
y_pred_lasso=L.predict(x_test)
```

```python
(mean_squared_error(y_test,y_pred_lasso))**0.5
```

```python
r2_score(y_test,y_pred_lasso)
```

```python
from sklearn.linear_model import ElasticNet
E=ElasticNet(l1_ratio=.8)
```

```python
E.fit(x_train,y_train)
y_pred_elas=E.predict(x_test)
```

```python
(mean_squared_error(y_test,y_pred_elas))**0.5
```

```python
r2_score(y_test,y_pred_elas)
```
