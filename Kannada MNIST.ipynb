```python
import tensorflow as tf
from tensorflow import keras
```

```python
x = df2.values[:, 1:].reshape((-1, 28, 28, 1)) / 255.
y = tf.keras.utils.to_categorical(df2.values[:, 0], 10)
a = df3.values[:, 1:].reshape((-1, 28, 28, 1)) / 255.
```

```python
from sklearn.model_selection import train_test_split as tts
from keras.models import Sequential
from keras.layers import Dense
from sklearn.metrics import accuracy_score
```

```python
x_train,x_test,y_train,y_test=tts(x,y,test_size=0.25,random_state=42)
```

```python
x_train.shape,y_train.shape
```

```python
model = tf.keras.Sequential([
                            tf.keras.layers.Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_uniform', padding='same', input_shape=(28,28,1)),
                            tf.keras.layers.BatchNormalization(),
                            tf.keras.layers.Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_uniform', padding='same'),
                            tf.keras.layers.MaxPooling2D(2,2),
                            tf.keras.layers.BatchNormalization(),
                            tf.keras.layers.Dropout(0.2),
                            tf.keras.layers.Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_uniform', padding='same'), 
                            tf.keras.layers.BatchNormalization(),
                            tf.keras.layers.Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_uniform', padding='same'),
                            tf.keras.layers.BatchNormalization(),
                            tf.keras.layers.MaxPooling2D(2,2),
                            tf.keras.layers.BatchNormalization(),
                            tf.keras.layers.Dropout(0.2),
                            tf.keras.layers.Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_uniform', padding='same'),
                            tf.keras.layers.BatchNormalization(),
                            tf.keras.layers.Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_uniform', padding='same'),
                            tf.keras.layers.BatchNormalization(),
                            tf.keras.layers.MaxPooling2D(2,2),
                            tf.keras.layers.Dropout(0.2), 
                            tf.keras.layers.BatchNormalization(),
                            tf.keras.layers.Flatten(),
                            tf.keras.layers.Dense(64, activation = 'relu', kernel_initializer='he_uniform'),
                            tf.keras.layers.BatchNormalization(),
                            tf.keras.layers.Dropout(0.2),
                            tf.keras.layers.Dense(10, activation = 'softmax'),])
                           

model.compile(optimizer ='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,validation_split=0.1,epochs=10)
```

```python
accuracy=model.evaluate(x_test,y_test)
```

```python
results = model.predict(a)
results = np.argmax(results,axis = 1)
results = pd.Series(results,name="label")
```

```python
submission = pd.concat([pd.Series(range(1,5001),name = "id"),results],axis = 1)
submission.to_csv("submission.csv",index=False)
```
