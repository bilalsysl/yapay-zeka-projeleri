```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```python
df1=pd.read_csv('train.csv')
df2=pd.read_csv('test.csv')
```

```python
address_encoding = {
    "Street": 0,
     "St": 0,
     "Avenue": 1,
     "Ave": 1,
     "Boulevard": 2,
     "Road": 3,
     "Drive": 4,
     "Lane": 5,
     "Tunnel": 6,
     "Highway": 7,
     "Way": 8,
     "Parkway": 9,
     "Parking": 10,
     "Oval": 11,
     "Square": 12,
     "Place": 13,
     "Bridge": 14
}
```

```python
def encode(x):
    if pd.isna(x):
        return None
    for road in address_encoding:
        if (road in x):
            return address_encoding[road]
    return None
```

```python
df1["EntryAdressEncoded"] = df1['EntryStreetName'].apply(encode)
df1['ExitAddressEncoded'] = df1['ExitStreetName'].apply(encode)
df2["EntryAdressEncoded"] = df2['EntryStreetName'].apply(encode)
df2['ExitAddressEncoded'] = df2['ExitStreetName'].apply(encode)
```

```python
pd.set_option('display.max_rows', 200)
```

```python
df1
```

```python
def adreskontrol(entry_address, exit_address):
    if (entry_address == exit_address):
        return True
    return False
```

```python
df1['IfEntryExitSame'] = df1.apply(lambda x: adreskontrol(x.EntryStreetName, x.ExitStreetName), axis=1)
df2['IfEntryExitSame'] = df2.apply(lambda x: adreskontrol(x.EntryStreetName, x.ExitStreetName), axis=1)
```

```python
directions = {
    'N': 0,
    'NE': 1/4,
    'E': 1/2,
    'SE': 3/4,
    'S': 1,
    'SW': 5/4,
    'W': 3/2,
    'NW': 7/4
}
```

```python
df1['EntryHeading'] = df1['EntryHeading'].map(directions)
df1['ExitHeading'] = df1['ExitHeading'].map(directions)

df2['EntryHeading'] = df2['EntryHeading'].map(directions)
df2['ExitHeading'] = df2['ExitHeading'].map(directions)
```

```python
df1.info()
```

```python
df1["Intersection"] = df1["IntersectionId"].astype(str) + df1["City"]
df2["Intersection"] = df2["IntersectionId"].astype(str) + df2["City"]
pd.concat([df1["Intersection"],df2["Intersection"]],axis=0).drop_duplicates().values
```

```python
le = preprocessing.LabelEncoder()
```

```python
le.fit(pd.concat([df1["Intersection"],df2["Intersection"]]).drop_duplicates().values)
df1["Intersection"] = le.transform(df1["Intersection"])
df2["Intersection"] = le.transform(df2["Intersection"])
```

```python
df2_dummy=pd.get_dummies(df2['City'])
```

```python
df2 = df2.join(df2_dummy)
```

```python
df1=df1.join(df2_dummy)
```

```python
df1.drop('City',axis=1,inplace=True)
df2.drop('City',axis=1,inplace=True)
```

```python
df1 = df1.dropna()
df2=df2.dropna()
```

```python
y = df1[["TotalTimeStopped_p20", "TotalTimeStopped_p50", "TotalTimeStopped_p80", "DistanceToFirstStop_p20","DistanceToFirstStop_p50", "DistanceToFirstStop_p80" ]]
x = df1[[ 'IntersectionId', "Intersection", 'Latitude', 'Longitude',
         'EntryHeading', 'ExitHeading', 'Hour', 'Weekend',
       'Month', 'EntryAdressEncoded', 'ExitAddressEncoded', 'IfEntryExitSame',
       'Atlanta', 'Boston', 'Chicago', 'Philadelphia']]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25 ,random_state=42)
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```
