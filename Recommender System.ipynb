```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('PBL 5 recommendation data.csv',encoding='latin-1')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.dropna(axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
df
```

```python
x=df[['Order_Items.product_name','Orders.order_number','Order_Items.price','Orders.status']]
```

```python
x
```

```python
x_grouped = x.groupby("Order_Items.product_name").agg({"Orders.status": [np.size, np.sum, np.mean]})
```

```python
x_grouped
```

```python
popular = x_grouped.sort_values(("Orders.status", "mean"), ascending= False)
```

```python
popular
```

```python
grouped_sum = popular["Orders.status","sum"].sum()
grouped_sum
```

```python
popular['percentage']=popular["Orders.status", "sum"].div(grouped_sum)*100
```

```python
popular
```

```python
popular["Rank"] = popular["percentage"].rank(ascending= False, method="min")
```
